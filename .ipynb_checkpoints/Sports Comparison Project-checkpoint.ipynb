```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Tweet Counts
import requests
import json

# Import API key
from api_keys import twitter_key

# Set API Token
bearer_token = twitter_key

# Set Search URL
search_url = "https://api.twitter.com/2/tweets/counts/recent"

# Twitter API format OAuth request
def bearer_oauth(r):
    """
    Method required by bearer token authentication.
    """

    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2RecentTweetCountsPython"
    return r

# Twitter API connect to server
def connect_to_endpoint(url, params):
    response = requests.request("GET", search_url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
def main():
    
    # Overview
        # Read leagues and corresponding team names from excel spreadsheet
        # Make API Requests
        # Clean data
        # Store league data in separate Data Frames
        # Compare leagues
    
    # Organize league data
    
    # Premier League Data
    # Read team names and twitter accounts from excel
    premier_league_data = pd.read_excel(open('Sports.xlsx', 'rb'),
              sheet_name="Premier League")
    # Set global variable
    global premier_league 
    # Create DataFrame
    premier_league = pd.DataFrame(premier_league_data, columns=["Team", "Account", "Start Date", "End Date", "Day 0 Tweets", "Day 1 Tweets", "Day 2 Tweets", "Day 3 Tweets", "Day 4 Tweets", "Day 5 Tweets", "Day 6 Tweets", "Day 7 Tweets", "Total Tweets"])
    
    # NFL
    nfl_data = pd.read_excel(open('Sports.xlsx', 'rb'),
              sheet_name='NFL')
    global nfl
    nfl = pd.DataFrame(nfl_data, columns=["Team", "Account", "Start Date", "End Date", "Day 0 Tweets", "Day 1 Tweets", "Day 2 Tweets", "Day 3 Tweets", "Day 4 Tweets", "Day 5 Tweets", "Day 6 Tweets", "Day 7 Tweets", "Total Tweets"])

    # MLB
    mlb_data = pd.read_excel(open('Sports.xlsx', 'rb'),
              sheet_name='MLB') 
    global mlb
    mlb = pd.DataFrame(mlb_data, columns=["Team", "Account", "Start Date", "End Date", "Day 0 Tweets", "Day 1 Tweets", "Day 2 Tweets", "Day 3 Tweets", "Day 4 Tweets", "Day 5 Tweets", "Day 6 Tweets", "Day 7 Tweets", "Total Tweets"])

    # NBA
    nba_data = pd.read_excel(open('Sports.xlsx', 'rb'),
              sheet_name='NBA')
    global nba
    nba = pd.DataFrame(nba_data, columns=["Team", "Account", "Start Date", "End Date", "Day 0 Tweets", "Day 1 Tweets", "Day 2 Tweets", "Day 3 Tweets", "Day 4 Tweets", "Day 5 Tweets", "Day 6 Tweets", "Day 7 Tweets", "Total Tweets"])

    # Loop through the Premier League
    for team in premier_league["Team"]:

        # Make API Request
        
        # Select current team's twitter account name from league dataframe
        account = premier_league.loc[premier_league["Team"] == team, "Account"].item()
        account = account.replace(u'\xa0', u' ')
        
        # Format parameters for request
        query_params = {'query': account, 'granularity': 'day'}
        
        # Make API Call
        json_response = connect_to_endpoint(search_url, query_params)
        
        # Add data to the league data frame
        
        # Data includes start date, end data, tweet count for each day, overall tweet count
        premier_league.loc[premier_league["Team"] == team, "Start Date"] = json_response["data"][0]["start"]
        premier_league.loc[premier_league["Team"] == team, "End Date"] = json_response["data"][7]["end"]
        premier_league.loc[premier_league["Team"] == team, "Day 0 Tweets"] = json_response["data"][0]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 1 Tweets"] = json_response["data"][1]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 2 Tweets"] = json_response["data"][2]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 3 Tweets"] = json_response["data"][3]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 4 Tweets"] = json_response["data"][4]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 5 Tweets"] = json_response["data"][5]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 6 Tweets"] = json_response["data"][6]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Day 7 Tweets"] = json_response["data"][7]["tweet_count"]
        premier_league.loc[premier_league["Team"] == team, "Total Tweets"] = premier_league.loc[premier_league["Team"] == team, "Day 0 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 1 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 2 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 3 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 4 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 5 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 6 Tweets"] + premier_league.loc[premier_league["Team"] == team, "Day 7 Tweets"]

    # Loop through the NFL
    for team in nfl["Team"]:
        
        # Make API Request
        account = nfl.loc[nfl["Team"] == team, "Account"].item()
        account = account.replace(u'\xa0', u' ')
        query_params = {'query': account, 'granularity': 'day'}        
        json_response = connect_to_endpoint(search_url, query_params)

        # Add data to the league data frame
        
        # Data includes start date, end data, tweet count for each day, overall tweet count  
        nfl.loc[nfl["Team"] == team, "Start Date"] = json_response["data"][0]["start"]
        nfl.loc[nfl["Team"] == team, "End Date"] = json_response["data"][7]["end"]
        nfl.loc[nfl["Team"] == team, "Day 0 Tweets"] = json_response["data"][0]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 1 Tweets"] = json_response["data"][1]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 2 Tweets"] = json_response["data"][2]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 3 Tweets"] = json_response["data"][3]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 4 Tweets"] = json_response["data"][4]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 5 Tweets"] = json_response["data"][5]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 6 Tweets"] = json_response["data"][6]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Day 7 Tweets"] = json_response["data"][7]["tweet_count"]
        nfl.loc[nfl["Team"] == team, "Total Tweets"] = nfl.loc[nfl["Team"] == team, "Day 0 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 1 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 2 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 3 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 4 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 5 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 6 Tweets"] + nfl.loc[nfl["Team"] == team, "Day 7 Tweets"]

    # Loop through the MLB
    for team in mlb["Team"]:

        # Make API Request
        account = mlb.loc[mlb["Team"] == team, "Account"].item()
        account = account.replace(u'\xa0', u' ')
        query_params = {'query': account, 'granularity': 'day'}            
        json_response = connect_to_endpoint(search_url, query_params)

        # Add data to the league data frame
        
        # Data includes start date, end data, tweet count for each day, overall tweet count
        mlb.loc[mlb["Team"] == team, "Start Date"] = json_response["data"][0]["start"]
        mlb.loc[mlb["Team"] == team, "End Date"] = json_response["data"][7]["end"]
        mlb.loc[mlb["Team"] == team, "Day 0 Tweets"] = json_response["data"][0]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 1 Tweets"] = json_response["data"][1]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 2 Tweets"] = json_response["data"][2]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 3 Tweets"] = json_response["data"][3]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 4 Tweets"] = json_response["data"][4]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 5 Tweets"] = json_response["data"][5]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 6 Tweets"] = json_response["data"][6]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Day 7 Tweets"] = json_response["data"][7]["tweet_count"]
        mlb.loc[mlb["Team"] == team, "Total Tweets"] = mlb.loc[mlb["Team"] == team, "Day 0 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 1 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 2 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 3 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 4 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 5 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 6 Tweets"] + mlb.loc[mlb["Team"] == team, "Day 7 Tweets"]

    # Loop through the NBA
    for team in nba["Team"]:

        # Make API Request
        account = nba.loc[nba["Team"] == team, "Account"].item()
        account = account.replace(u'\xa0', u' ')
        query_params = {'query': account, 'granularity': 'day'}    
        json_response = connect_to_endpoint(search_url, query_params)

        # Add data to the league data frame
        
        # Data includes start date, end data, tweet count for each day, overall tweet count
        nba.loc[nba["Team"] == team, "Start Date"] = json_response["data"][0]["start"]
        nba.loc[nba["Team"] == team, "End Date"] = json_response["data"][7]["end"]
        nba.loc[nba["Team"] == team, "Day 0 Tweets"] = json_response["data"][0]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 1 Tweets"] = json_response["data"][1]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 2 Tweets"] = json_response["data"][2]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 3 Tweets"] = json_response["data"][3]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 4 Tweets"] = json_response["data"][4]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 5 Tweets"] = json_response["data"][5]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 6 Tweets"] = json_response["data"][6]["tweet_count"]
        nba.loc[nba["Team"] == team, "Day 7 Tweets"] = json_response["data"][7]["tweet_count"]
        nba.loc[nba["Team"] == team, "Total Tweets"] = nba.loc[nba["Team"] == team, "Day 0 Tweets"] + nba.loc[nba["Team"] == team, "Day 1 Tweets"] + nba.loc[nba["Team"] == team, "Day 2 Tweets"] + nba.loc[nba["Team"] == team, "Day 3 Tweets"] +nba.loc[nba["Team"] == team, "Day 4 Tweets"] + nba.loc[nba["Team"] == team, "Day 5 Tweets"] +nba.loc[nba["Team"] == team, "Day 6 Tweets"] + nba.loc[nba["Team"] == team, "Day 7 Tweets"]
      
if __name__ == "__main__":
    main()
```
